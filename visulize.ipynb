```python
import matplotlib.pyplot as plt
import pandas as pd

aircraft_df = pd.read_csv("aircraft.csv")
country_df = pd.read_csv("country.csv")
# print(aircraft_df)
print(country_df)
```

```python
x = aircraft_df['   Aircraft_family_code']
x1 = x.iloc[:9]
y = aircraft_df['   No_of_aircrafts']
y1 = y.iloc[:9]


plt.bar(x1,y1)
plt.title("Aircraft Detail")
plt.xlabel("Aircraft_family_code")
plt.ylabel("No of Aircraft")
plt.show()
```

```python

x2 = x.iloc[9:]
y2 = y.iloc[9:]

plt.bar(x2,y2)
plt.title("Aircraft Detail")
plt.xlabel("Aircraft_family_code")
plt.ylabel("No of Aircraft")
plt.show()
```

```python
fig, axe = plt.subplots(2,1)

axe[0].bar(x1,y1)
axe[0].set_title("Aircrafts detail")
axe[0].set_ylabel("no of aircrafts")

axe[1].bar(x2,y2)
axe[1].set_xlabel("Aircraft family")
axe[1].set_ylabel("no of aircrafts")
```
